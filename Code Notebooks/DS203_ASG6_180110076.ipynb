# Q1

```
import pandas as pd
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
import seaborn as sn
```

```
mdf=pd.read_csv("master.csv")
mdf.head(10)
```

```
# Top 6 populous countries with average annual population
cydf=mdf.groupby(["country","year"])
py=cydf["population"].sum().to_frame()
cp=py.mean(level=0)
top=cp.sort_values("population",ascending=False)[:6]
top
```

```
# Annual data on these countries
cdf=mdf.groupby("country")
top_list=["Russian Federation","United States","Japan","Mexico","Germany","Brazil"]
means=[]
variances=[]
for country in top_list:
  print("\033[1m"+country+"\033[0m")
  df=((cdf.get_group(country)).groupby("year")).sum()
  df["suicides/100k pop"]=df["suicides_no"]*100000/df["population"]
  display(df)
  avg=np.mean(df["suicides/100k pop"].values)
  means.append(avg)
  print("Average number of suicides per 100,000 in a year  = "+str(avg))
  var=np.var(df["suicides/100k pop"].values)
  variances.append(var)
  print("Sample variance for the same = "+str(var))
  print()
```

```
# QQ plots
for country in top_list:
  print("\033[1m"+country+"\033[0m")
  df=((cdf.get_group(country)).groupby("year")).sum()
  df["suicides/100k pop"]=df["suicides_no"]*100000/df["population"]
  ss.probplot(df["suicides/100k pop"],plot=plt)
  plt.show()
```

```
# Log-likelihood values
for i in range(6):
  country=top_list[i]
  mean=means[i]
  var=variances[i]
  print("\033[1m"+country+"\033[0m")
  df=((cdf.get_group(country)).groupby("year")).sum()
  df["suicides/100k pop"]=df["suicides_no"]*100000/df["population"]
  print(np.log(ss.norm.pdf(df["suicides/100k pop"],mean,var)).sum())
```

Greater log-likelihood values mean QQ plot is closer to linear fit and the underlying distribution is more likely to be gaussian<br>
This seems to be the case for US, Mexico & Brazil from both QQ plots & log likelihoods

```
for i in range(6):
  country=top_list[i]
  print("\033[1m"+country+"\033[0m")
  df=((cdf.get_group(country)).groupby("year")).sum()
  df["suicides/100k pop"]=df["suicides_no"]*100000/df["population"]
  data=df['suicides/100k pop']
  SE=ss.sem(data)
  zscore=ss.norm.ppf(0.975)
  h=SE*zscore
  m=means[i]
  start=m-h
  end=m+h
  print("95% CI for mean of yearly suicides/100k pop = ("+str(start)+","+str(end)+")")
```

Mexico & Brazil have closest means over the same set of years (1985-2015)

```
# Welch's t-test & Wilcoxon signed rank test
BZ=((cdf.get_group("Brazil")).groupby("year")).sum()
BZ["suicides/100k pop"]=BZ["suicides_no"]*100000/BZ["population"]
MX=((cdf.get_group("Mexico")).groupby("year")).sum()
MX["suicides/100k pop"]=MX["suicides_no"]*100000/MX["population"]
print(ss.ttest_ind(BZ["suicides/100k pop"],MX["suicides/100k pop"],equal_var=False))
print(ss.wilcoxon(BZ["suicides/100k pop"],MX["suicides/100k pop"]))
```

p-value is very low (below 1% in both tests), hence the means are indeed different

```
# Paired t-test
ss.ttest_rel(BZ["suicides/100k pop"],MX["suicides/100k pop"])
```

The p-value is extremely low or negligible for the test when data is paired by year

```
sy=cydf["suicides_no"].sum().to_frame()
py=cydf["population"].sum().to_frame()
hy=cydf.first()["HDI for year"].to_frame()
gy=cydf.first()["gdp_per_capita ($)"].to_frame()
ydf=pd.concat([sy,py,hy,gy],axis=1)
ydf["suicides/100k pop"]=ydf["suicides_no"]*100000/ydf["population"]
ydf
```

```
ydf.corr(method="pearson")
```

All correlations are positive: suicide rate increases with increase in HDI & per capita GDP but with weak relationship

# Q2

```
prodf=pd.read_csv("summer-products-with-rating-and-performance_2020-08.csv")
prodf["revenue"]=prodf["price"]*prodf["units_sold"]
prodf["discount"]=prodf["retail_price"]-prodf["price"]
prodf["discount%"]=(prodf["retail_price"]-prodf["price"])*100/prodf["retail_price"]
prodf.head()
```

```
for col in ["rating","merchant_rating","discount%"]:
  data=prodf[col]
  avg=np.mean(data.values)
  var=np.var(data.values)
  print("Mean of "+col+" = "+str(avg))
  print("Variance of "+col+" = "+str(var))
  print()
```

```
OC=prodf.groupby("origin_country")
for name, group in OC:
  print(name)
  display(group)
  print()
```

```
for country in ["CN","US","VE"]:
  print("\033[1m"+country+"\033[0m")
  df=OC.get_group(country)
  for col in ["rating","merchant_rating","discount%"]:
    data=df[col]
    avg=np.mean(data.values)
    var=np.var(data.values)
    print("Sample Mean of "+col+" = "+str(avg))
    print("Sample Variance of "+col+" = "+str(var))
    SE=ss.sem(data)
    zscore=ss.norm.ppf(0.975)
    h=SE*zscore
    start=avg-h
    end=avg+h
    print("95% CI for mean = ("+str(start)+","+str(end)+")")
    ss.probplot(data,plot=plt)
    plt.show()
    print()
```

```
ad=prodf.groupby("uses_ad_boosts")
for name, group in ad:
  print(name)
  display(group)
  print()
```

```
NoAd=ad.get_group(0)
YesAd=ad.get_group(1)
m1=np.mean(NoAd["units_sold"])
m2=np.mean(YesAd["units_sold"])
print("Mean units sold without using ad boosts= "+str(int(m1)))
print("Mean units sold when using ad boosts= "+str(int(m2)))
```

Welch t-test to test equality of means of units_sold for cases of use/non-use of ad boosts<br>
Let the null hypothesis be that use of ad boosts does not affect sales, it would be rejected if p<0.01

```
u1=NoAd["units_sold"]
u2=YesAd["units_sold"]
print(ss.ttest_ind(u1,u2,equal_var=False))
```

p-value of 0.52 indicates null hypothesis should be accepted i.e. statistically no effect of using ad boosts on sales

```
r1=NoAd["revenue"]
r2=YesAd["revenue"]
print(ss.ttest_ind(r1,r2,equal_var=False))
```

Statistically no effect visible of using ad boosts on revenues

```
bg=prodf.groupby("badges_count")
for name, group in bg:
  print(name)
  display(group)
  print()
```

```
bno=bg.get_group(0)
b1=bg.get_group(1)
b2=bg.get_group(2)
b3=bg.get_group(3)
byes=pd.concat([b1,b2,b3])
m1=np.mean(bno["merchant_rating"])
m2=np.mean(byes["merchant_rating"])
print("Mean ratings when merchant has no badge = "+str(m1))
print("Mean ratings when merchant has at least one badge = "+str(m2))
```

Welch t-test to test equality of means of merchant ratings in cases of having at least one/not having a badge<br>
Let the null hypothesis be that possessing a badge doesn't affect ratings of merchant, it would be rejected if p<0.01

```
mr1=bno["merchant_rating"]
mr2=byes["merchant_rating"]
print(ss.ttest_ind(mr1,mr2,equal_var=False))
```

Extremely low p-value: null hypothesis is rejected. The data strongly indicates that a merchant awarded with badges receives higher ratings.

```
br1=bno["rating"]
br2=byes["rating"]
print(ss.ttest_ind(br1,br2,equal_var=False))
```

Products with badges have higher ratings

```
f3=prodf["uses_ad_boosts"].sum()
f4=prodf.shape[0]-f1
print(f3,f4)
```

Does the proportion of products using & not using ad boosts from US or CN a representative of global? <br>
Null hypothesis: Yes, it is. Reject if p<0.01

```
CN=OC.get_group("CN")
f1=CN["uses_ad_boosts"].sum()
f2=CN.shape[0]-f1
print(f1,f2)
ss.chisquare(f_obs=[f1,f2],f_exp=[f3,f4])
```

p>0.01 So accept null hypothesis i.e products from CN have same proportion of ad boosts as from anywhere

```
US=OC.get_group("US")
f1=US["uses_ad_boosts"].sum()
f2=US.shape[0]-f1
print(f1,f2)
ss.chisquare(f_obs=[f1,f2],f_exp=[f3,f4])
```

p=0! Null hypothesis is rejected. US isn't representative of proportion of products using ad boosts at global level.
