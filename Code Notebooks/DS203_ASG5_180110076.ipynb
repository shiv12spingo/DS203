# Q1

```
import pandas as pd
mdf=pd.read_csv("master.csv")
mdf.head(10)
```

```
cdf=mdf.groupby("country")
cdf.first()
```

Initial record for each country. Total 101 countries.

```
# Displaying first record for each year in each country
cydf=mdf.groupby(["country","year"])
cydf.first()
```

```
# Total number of suicides every year for each country
sy=cydf["suicides_no"].sum().to_frame()
sy
```

```
# Average number of suicides in a year for each country
cs=sy.mean(level=0)
cs
```

```
# Top 6 countries by average yearly suicides
top=cs.sort_values("suicides_no",ascending=False)[:6]
top
```

```
# Number of suicides in each year for each of these 6 countries
top_list=["Russian Federation","United States","Japan","Ukraine","Germany","France"]
for country in top_list:
  print("\033[1m"+country+"\033[0m")
  display(((cdf.get_group(country)).groupby("year")).sum()["suicides_no"].to_frame())
  print()
```

```
import matplotlib.pyplot as plt
import seaborn as sn
top=mdf[mdf["country"].isin(top_list)]
fig,ax=plt.subplots(figsize=(20,15))
ax=sn.barplot(x=top['country'], y='suicides/100k pop', hue='age', data=top, palette='rocket')
```

In all the 6 countries majorly contributing to suicide numbers, its clear that as age increases the tendency of suicide increases as greater proportion of people (no. of suicides per 100k population) are found to suicide with inceasing age.

```
fig,ax=plt.subplots(figsize=(20,15))
ax=sn.barplot(x=top['country'], y='suicides/100k pop', hue='sex', data=top, palette='rocket')
```

```
fig,ax=plt.subplots(figsize=(20,15))
ax=sn.barplot(x=mdf['age'], y='suicides/100k pop', hue='sex', data=top, palette='rocket')
```

```
fig,ax=plt.subplots(figsize=(20,15))
ax=sn.barplot(x=mdf['generation'], y='suicides/100k pop', hue='sex', data=top, palette='rocket')
```

The tendency of suicide is clearly much larger in males as compared to females (comparing their relative proportions) and this is visible across generations, age-groups and major contributing countries.

```
usgen=(cdf.get_group('United States')).groupby(["generation","year"])["suicides_no"].sum().to_frame()
usgen
```

```
ussdf=usgen.reset_index().groupby("generation")
for gen in mdf["generation"].unique():
  data=ussdf.get_group(gen)
  x=data["year"]
  y=data["suicides_no"]
  ymax=max(y)
  xmax=data[y==ymax]["year"]
  fig,ax=plt.subplots(figsize=(15,7))
  ax.plot(x,y)
  ax.plot(xmax,ymax,'ro')
  ax.annotate(s="Worst Year = "+str(int(xmax)), xy=(xmax,ymax), xytext=(xmax+0.5,ymax))
  ax=plt.xlabel("Year")
  ax=plt.ylabel("Annual number of suicides")
  ax=plt.title(gen)
```

Variation of annual suicides with year for different generations with worst year being the one with maximum number of suicides

```
US=cdf.get_group("United States")
male=US[US["sex"]=="male"]
m2020=male[male["year"]==2000]
m=m2020[["age","population"]]
m
```

```
female=US[US["sex"]=="female"]
f2020=female[female["year"]==2000]
f=f2020[["age","population"]]
f
```

```
fig,ax=plt.subplots(figsize=(20,10))
ax.barh(y=m["age"], width=m["population"])
ax.barh(y=f["age"], width=-f["population"])
ax.legend(["Male","Female"])
ax.set_ylabel("Age Group")
ax.set_xlabel("Population/10,000,000")
ax.set_title("Sex wise Population across different age groups in US in the year 2000")
labels = [item.get_text() for item in ax.get_xticklabels()]
labels = ["6","4","2","0","2","4","6"]
ax.set_xticklabels(labels)
```

```
G=cdf.get_group("Germany")
male=G[G["sex"]=="male"]
m2000=male[male["year"]==2000]
mG=m2000[["age","population"]]
fmale=G[G["sex"]=="female"]
f2000=fmale[fmale["year"]==2000]
fG=f2000[["age","population"]]
```

```
fig,ax=plt.subplots(figsize=(20,10))
ax.barh(y=mG["age"], width=mG["population"])
ax.barh(y=fG["age"], width=-fG["population"])
ax.legend(["Male","Female"])
ax.set_ylabel("Age Group")
ax.set_xlabel("Population in 10 million")
ax.set_title("Sex wise Population across different age groups in Germany in the year 2000")
labels = ["1.5","1","0.5","0","0.5","1","1.5"]
ax.set_xticklabels(labels)
```

```
py=cydf["population"].sum().to_frame()
hy=cydf.first()["HDI for year"].to_frame()
bubdf=pd.concat([sy,py,hy],axis=1)
bubdf
```

```
import plotly.express as px
fig = px.scatter(bubdf, x="HDI for year",y="suicides_no", size="population",
                 labels={"suicides_no":"No. of suicides in the year"},
                 title="Variation of no. of suicides with HDI for all years with differing populations of countries")
fig.show()
```

If we compare positions of bubbles with same large sizes it seems that for countries with significant population:<br> The number of suicides in a year increase with increase in HDI

```
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
df=pd.read_csv("summer-products-with-rating-and-performance_2020-08.csv")
```

```
df["revenue"]=df["price"]*df["units_sold"]
df["discount"]=df["retail_price"]-df["price"]
df["discount%"]=(df["retail_price"]-df["price"])*100/df["retail_price"]
df.head(5)
```

```
fig,ax=plt.subplots(figsize=(20,10))
CM=df[["price","shipping_option_price","retail_price","discount","discount%",
       "badges_count","badge_local_product","badge_product_quality","badge_fast_shipping",
       "units_sold","revenue","rating","rating_count","merchant_rating","merchant_rating_count"]].corr()
ax=sn.heatmap(CM,annot=True)
```

```
# Top 50 revenue generating products, displaying top 5
top_revenue=df.sort_values("revenue",ascending=False)[:50]
top_revenue.head()
```

```
fig,ax=plt.subplots(figsize=(10,5))
ax=sn.histplot(x="discount%",data=df)
```

```
fig,ax=plt.subplots(figsize=(10,5))
ax=sn.distplot(a=df["rating"])
```

Most products have mean rating lying in range 3 to 4.5

```
fig,ax=plt.subplots(figsize=(10,5))
ax=sn.distplot(a=df["merchant_rating"])
```

Most merchants get rating mostly around 4

```
# Mean Ratings for products with different number of badges
fig,ax=plt.subplots(figsize=(6,6))
ax=sn.barplot(x="badges_count", y='rating', data=df)
```

```
# Mean no of units sold for products with different number of badges
fig,ax=plt.subplots(figsize=(6,8))
ax=sn.barplot(x="badges_count", y='units_sold', data=df)
```

```
fig,ax=plt.subplots(figsize=(4,4))
ax=sn.barplot(x="badge_local_product", y='rating', data=df)
```

Whether a product is local or not does not influence rating

```
fig,ax=plt.subplots(figsize=(4,4))
ax=sn.barplot(x="badge_local_product", y='units_sold', data=df)
```

A local product badge is not compelling enough for increasing sales

```
fig,ax=plt.subplots(figsize=(3,5))
ax=sn.barplot(x="badge_product_quality", y='rating', data=df)
```

Product with a badge of good product quality has better ratings

```
fig,ax=plt.subplots(figsize=(3,6))
ax=sn.barplot(x="badge_product_quality", y='units_sold', data=df)
```

Better quality product badge has resulted in greater sales

```
fig,ax=plt.subplots(figsize=(4,5))
ax=sn.barplot(x="uses_ad_boosts", y='revenue', data=df)
```

Use of ad boosts has not resulted in increased revenues

```
# Mean Revenues for products with different countries as origin
fig,ax=plt.subplots(figsize=(8,8))
ax=sn.barplot(x="origin_country", y='revenue', data=df)
```

```
# Mean revenues for different discount% offered
df["discount%_range"]=pd.cut(df["discount%"],bins=np.arange(-20, 120, 20))
fig,ax=plt.subplots(figsize=(8,8))
ax=sn.barplot(x="discount%_range", y='revenue', data=df)
```

```
# Mean number of units sold for different ratings
df["rating_range"]=pd.cut(df["rating"],bins=np.arange(0,6,1))
fig,ax=plt.subplots(figsize=(7,7))
ax=sn.barplot(x="rating_range", y='units_sold', data=df)
```

Products with poor ratings (0-3) don't sell much

```
fig,ax=plt.subplots(figsize=(15,8))
ax=sn.barplot(x="discount%_range", y='units_sold', hue="rating_range", data=df, palette="rocket")
```

```
fig,ax=plt.subplots(figsize=(12,8))
ax=sn.barplot(x="discount%_range", y='units_sold', hue="badge_product_quality", data=df, palette="rocket")
```

Quality (1) and ratings (4 or 5) do have positive impact on sales but that of discounts isn't evident
