# Q1

```
import pandas as pd
import numpy as np
```

```
df=pd.read_excel("ENB2012_data.xlsx")
df
```

```
X = df.drop(["Y1","Y2"],axis=1)
Y = df[["Y1","Y2"]]
print(X.shape,Y.shape)
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)
print(X_train.shape,Y_train.shape)
```

```
from sklearn.linear_model import LinearRegression
LM = LinearRegression()
LM.fit(X_train,Y_train)
```

```
LM.coef_
```

```
LM.intercept_
```

```
Y_pred = LM.predict(X_test)
Y_pred = pd.DataFrame(Y_pred,columns=["Y1","Y2"])
Y_pred
```

```
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(Y_pred,Y_test)
r2 = r2_score(Y_test,Y_pred)
print(mse,r2)
```

```
for test_split in [0.4,0.3,0.2,0.1]:
  print("Test Split = "+str(test_split))
  X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=test_split)
  LM.fit(X_train,Y_train)
  print(f"Model coefficients: {LM.coef_.tolist()}")
  print(f"Intercepts: {LM.intercept_.tolist()}")
  Y_pred = LM.predict(X_test)
  display(pd.DataFrame(Y_pred,columns=["Y1","Y2"]))
  mse,r2 = mean_squared_error(Y_pred,Y_test),r2_score(Y_test,Y_pred)
  print("MSE = "+str(mse),"R2 = "+str(r2))
  print()
```

Least MSE (8.522) obtained for Test Split = 0.3 & Highest R2 (0.9067) obtained for Test Split = 0.1

```
from sklearn.linear_model import Lasso,Ridge
```

```
# Ridge Regression Model
for alpha in [0.001,0.005,0.01,0.05,0.1,0.5]:
  print("Alpha = "+str(alpha))
  for test_split in [0.4,0.3,0.2,0.1]:
    X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = test_split)
    RM = Ridge(alpha=alpha)
    RM.fit(X_train,Y_train)
    Y_pred = RM.predict(X_test)
    mse = mean_squared_error(Y_pred,Y_test)
    r2 = r2_score(Y_test,Y_pred)
    print(f'Test split: {test_split}  MSE: {mse}  R2: {r2}')
```

Least MSE = 8.11 & Best R2 = 0.924 are obtained for same Alpha = 0.005, Test Split = 0.1

```
# Lasso Regression Model
import warnings
warnings.filterwarnings("ignore") # Warnings are suppressed for better view of print statements
for alpha in [0.001,0.005,0.01,0.05,0.1,0.5]:
  print("Alpha = "+str(alpha))
  for test_split in [0.4,0.3,0.2,0.1]:
    X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = test_split)
    LsM = Lasso(alpha=alpha)
    LsM.fit(X_train,Y_train)
    Y_pred = LsM.predict(X_test)
    mse = mean_squared_error(Y_pred,Y_test)
    r2 = r2_score(Y_test,Y_pred)
    print(f'Test split: {test_split}  MSE: {mse}  R2: {r2}')
```

Least MSE = 5.53 & Best R2 = 0.94 are obtained for same Alpha = 0.001 & Test Split = 0.1

Lasso Regression Model has yielded least Mean Square Error and best value of R^2

```
X2_train,X2_test,Y2_train,Y2_test = train_test_split(X2,Y2,test_size = 0.2)
print(X2_train.shape,Y2_train.shape)
```

```
LgM.coef_
```

```
LgM.intercept_
```

```
Y2_pred = LgM.predict(X2_test)
Y2_pred_df = pd.DataFrame(Y2_pred,columns=["Status"])
Y2_pred_df
```

```
# Indicator function
def indicator(a,b):
  ind=[1 if a[i]!=b[i] else 0 for i in range(len(a))]
  return np.array(ind)
```

```
# Classification Error
ind = indicator(Y2_pred,np.array(Y2_test))
ce = (np.sum(ind))/len(Y2_pred)
ce
```

```
for test_split in [0.4,0.3,0.2,0.1]:
  print("Test Split = "+str(test_split))
  X2_train,X2_test,Y2_train,Y2_test = train_test_split(X2,Y2,test_size = 0.2)
  LgM.fit(X2_train,Y2_train)
  print(f"Model coefficients: {LgM.coef_.tolist()[0]}")
  print(f"Intercept = {LgM.intercept_.tolist()[0]}")
  Y2_pred = LgM.predict(X2_test)
  print(f"Predicted Statuses: {Y2_pred.tolist()}")
  ind = indicator(Y2_pred,np.array(Y2_test))
  ce = (np.sum(ind))/len(Y2_pred)
  print("Classification Error = "+str(ce))
  print()
```

Least classification error (0.258) obtained for Test Split = 0.3
